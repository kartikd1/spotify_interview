```python
df['monthly_active_user_pct'] = df.mau/df.users

df['monthly_active_user_pct']
```

#### ASSUMPTION: Proportion of monthly active users roughly should translate to the daily active user proportion
#### TODO: Think of cases where this can't be true

```python
df['daily_total_users'] = df.dau/df.monthly_active_user_pct
df['daily_total_users'] = df['daily_total_users'].apply(round)
```

```python
df[['mau', 'users', 'dau', 'monthly_active_user_pct', 'daily_total_users', 'streams', 'skippers']]
```

##### Metric: Streams per user

```python
df['streams_per_user'] = np.where(df['daily_total_users'] > 0, df['streams']/df['daily_total_users'], 0)

df['streams_per_user'] = df['streams_per_user'].apply(round)
```

```python
df.groupby('streams_per_user', as_index=False).count()
```

##### Metric: proportion of skippers per total user

```python
df['skipers_to_users'] = np.where(df.skippers != 0, df.skippers/df.daily_total_users, 0)

df.plot('skipers_to_users'
```

```python
df[(df.daily_total_users == 0) & (df.skippers > 0)]
```

### Creating a response variable based on proportion of skippers to DAU 

```python
df['success_flg_daily'] = np.where(df.skippers_to_active == -1, 'bad'
                                   , np.where(df.skippers_to_active == 0, 'good'
                                    , np.where((df.skippers_to_active > 0) & (df.skippers_to_active <= 0.5), 'good'
                                    , np.where(df.skippers_to_active >= 0.5, 'bad', 'missing')
                                                                               )))
```

```python
df.groupby('success_flg_daily', as_index=False)['playlist_uri'].count()
```

```python
sns.displot(data=df, x="skippers_to_mau", hue="success_flg_daily", height=8, aspect=12/8, kind='kde')
```
